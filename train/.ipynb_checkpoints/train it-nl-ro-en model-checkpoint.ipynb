## Creating train, valid and test sets

```python
# Excluding languages without 11000 lines

lang_list = []

for l in original_list_of_languages:
    with open(PMINDIA_CORPUS_DIR + "pmindia.v1.{}-en.tsv".format(l), encoding='utf8') as f:
        num_lines = sum(1 for line in f)
        if num_lines >= 11000:
            lang_list.append(l)
```

```python
lang_list
#lang_list = ['ta','hi']
```

```python
# Finding longest subsequence among lists

lists = []

for lang in lang_list:
    with open(PMINDIA_CORPUS_DIR + "pmindia.v1.{}-en.tsv".format(lang), encoding='utf8') as f:
        lists.append([l.split("\t")[0] for l in f])
```

```python
def find_sentence_pos(lists,l,sentence):
    try:
        i = lists[l].index(sentence)
    except:
        return None
    return i

def are_all_pos_equal(positions,lists):
    if len(lists[0]) <= positions[0]:
        return False
    sentence = lists[0][positions[0]]
    for i in range(1,len(lists)):
        if len(lists[i]) <= positions[i]:
            return False
        if not sentence == lists[i][positions[i]]:
            return False
    return True
```

```python
#find sequences
spans = []
print(len(lists[0]))
for i in range(len(lists[0])):
    if i % 1000 == 0:
        print(i)
    positions = [None for p in range(len(lists))]
    positions[0] = i
    for l in range(1,len(lists)):
        positions[l] = find_sentence_pos(lists,l,lists[0][i])
    if None in positions:
        continue
    condition = True
    length = 0
    while condition:
        if are_all_pos_equal(positions,lists):
            positions = [p+1 for p in positions]
            length +=1
        else:
            if length > 0:
                spans.append((i,length))
            condition = False
```

```python
from operator import itemgetter
longest_span = max(spans,key=itemgetter(1))
#print("{}, {}".format(longest_span[0],longest_span[0]+longest_span[1]))
print(longest_span)
```

```python
#find all elems present in all sets
alignments = []
print(len(lists[0]))
for i in range(len(lists[0])):
    if i % 1000 == 0:
        print(i)
    positions = [None for p in range(len(lists))]
    positions[0] = i
    for l in range(1,len(lists)):
        positions[l] = find_sentence_pos(lists,l,lists[0][i])
    if None in positions:
        continue
    alignments.append(positions)
```

### Tokenizing data

```python
from tokenizers import Tokenizer
from tokenizers.models import WordPiece

ind_en_tokenizer = Tokenizer(WordPiece(unk_token="[UNK]"))
```

```python
from tokenizers.trainers import WordPieceTrainer

ind_en_trainer = WordPieceTrainer(special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"], vocab_size=50000, show_progress=True)
```

```python
from tokenizers.pre_tokenizers import Whitespace

ind_en_tokenizer.pre_tokenizer = Whitespace()

en_bn_files = [EN_BN_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
bn_en_files = [EN_BN_CORPUS_DIR + f"{split}.bn" for split in ["valid","train","test"]]

en_gu_files = [EN_GU_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
gu_en_files = [EN_GU_CORPUS_DIR + f"{split}.gu" for split in ["valid","train","test"]]

en_hi_files = [EN_HI_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
hi_en_files = [EN_HI_CORPUS_DIR + f"{split}.hi" for split in ["valid","train","test"]]

en_kn_files = [EN_KN_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
kn_en_files = [EN_KN_CORPUS_DIR + f"{split}.kn" for split in ["valid","train","test"]]

en_ml_files = [EN_ML_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
ml_en_files = [EN_ML_CORPUS_DIR + f"{split}.ml" for split in ["valid","train","test"]]

en_mr_files = [EN_MR_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
mr_en_files = [EN_MR_CORPUS_DIR + f"{split}.mr" for split in ["valid","train","test"]]

en_or_files = [EN_OR_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
or_en_files = [EN_OR_CORPUS_DIR + f"{split}.or" for split in ["valid","train","test"]]

en_pa_files = [EN_PA_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
pa_en_files = [EN_PA_CORPUS_DIR + f"{split}.pa" for split in ["valid","train","test"]]

en_ta_files = [EN_TA_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
ta_en_files = [EN_TA_CORPUS_DIR + f"{split}.ta" for split in ["valid","train","test"]]

en_te_files = [EN_TE_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
te_en_files = [EN_TE_CORPUS_DIR + f"{split}.te" for split in ["valid","train","test"]]

en_ur_files = [EN_UR_CORPUS_DIR + f"{split}.en" for split in ["valid","train","test"]]
ur_en_files = [EN_UR_CORPUS_DIR + f"{split}.ur" for split in ["valid","train","test"]]

files = en_bn_files + bn_en_files + en_gu_files + gu_en_files + en_hi_files + hi_en_files + en_kn_files + kn_en_files + en_ml_files + ml_en_files + en_mr_files + mr_en_files + en_or_files + or_en_files + en_pa_files + pa_en_files + en_ta_files + ta_en_files + en_te_files +te_en_files + en_ur_files + ur_en_files
ind_en_tokenizer.train(files= files, trainer=ind_en_trainer)
```

```python
from pathlib import Path



def tokenize_files(tokenizer, files, extension, output_path):
    for file in files:
        print(f"Reading file {file}")
        with open(file, encoding='utf8') as f:
          lines = f.readlines()
          tokenized_lines = tokenizer.encode_batch(lines)
          tokenized_name = Path(file).stem
          tokenized_name = output_path + tokenized_name + "." + extension
          print(tokenized_name)
          with open(tokenized_name, 'w', encoding='utf8') as wf:

            wf.writelines([" ".join(t.tokens) + "\n" for t in tokenized_lines])
```
